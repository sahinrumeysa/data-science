```python
import matplotlib.pyplot as plt
import yfinance as yf

# Fetching Tesla stock data
tesla_data = yf.download('TSLA', start='2022-09-09', end='2023-09-09')

# Creating the plot
plt.plot(tesla_data.index, tesla_data['Close'])

# Setting the title of the plot
plt.title('Tesla Stock Graph')

# Setting labels for the axes
plt.xlabel('Date')
plt.ylabel('Closing Price')

# Displaying the plot
plt.show()
```
